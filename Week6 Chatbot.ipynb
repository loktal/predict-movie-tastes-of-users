```python
#2

def extractorLine(ligne):
    movieID=ligne[0]
    movieGenre=ligne[2]
    movieTitle=""
    movieYear=""
    mot=ligne[1]
    i=0
    while mot[i]!="(":
      movieTitle+=mot[i]
      i+=1

        
    for i in range(len(mot)):
      if mot[i]=="(":
        movieYear=mot[i+1]+mot[i+2]+mot[i+3]+mot[i+4]

    return int(movieID),movieTitle,int(movieYear),movieGenre
```

```python
def normalisation(ligne):
    
    movieID=ligne[0]
    
    movieTitle=""
    movieYear=""
    
    ##################### on divise titre et année #####################
    mot=ligne[1]
    i=0
    while mot[i]!="(":
      movieTitle+=mot[i]
      i+=1

        
    for i in range(len(mot)):
      if mot[i]=="(":
        movieYear=mot[i+1]+mot[i+2]+mot[i+3]+mot[i+4]
    
    ##################### on divise les genre #####################
    
    movieGenre=ligne[2]
    
    movieGenre=movieGenre.split("|")
    
    listeGenre= [0 for i in range(19)]
    
    
    for genre in movieGenre:
        for i in range(len(all_genres)):
            if genre == all_genres[i]:
                listeGenre[i]=1
                
                
    ##################### on normalize #####################
    count=0
    for item in listeGenre:
        count += item
    
    if(count != 0):
        for i in range(len(listeGenre)):
            listeGenre[i]=listeGenre[i]/count
    
    
    

    return int(movieID),movieTitle,int(movieYear),listeGenre[0],listeGenre[1],listeGenre[2],listeGenre[3],listeGenre[4],listeGenre[5],listeGenre[6],listeGenre[7],listeGenre[8],listeGenre[9],listeGenre[10],listeGenre[11],listeGenre[12],listeGenre[13],listeGenre[14],listeGenre[15],listeGenre[16],listeGenre[17],listeGenre[18]



```
